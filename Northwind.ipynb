## Evaluating Employee Sales

The management team wants to know who are their top-performing employees to reward them. They also want to know which employees are struggling to offer them the necessary training or resources they need to improve.

```python
%%sql
    -- 'Getting employee name, title, hire date, and total sales'
    WITH
    employee_sales AS
    (
    SELECT  e.first_name || ' ' || e.last_name AS employee, e.title, e.hire_date,
            ROUND(SUM(od.unit_price * od.quantity * (1-od.discount))::numeric,2) as total_sale
      FROM orders o
      JOIN order_details od
        ON o.order_id = od.order_id
      JOIN employees e
        ON o.employee_id = e.employee_id
     GROUP BY e.employee_id
    )

-- 'Ranking the employees based on total sales'
SELECT *, RANK() OVER(ORDER BY total_sale DESC)
FROM employee_sales;
```

It looks like Margaret Peacock has the highest amount of total sales amongest the sales representatives.  From the table it shows that the more senior employees have higher total sales but we can see Michael Suyama has the lowest total sales.  Northwind might want to come up with a development plan with Michael to help him improve his sales.

## Northwind Monthly Sales

Management wants to know how they are doing in monthly sales and how much the company has made currently.

```python
%%sql
    -- 'Grouping data into year and months then getting total sale for each month'
    WITH
    monthly_sales AS 
    (
    SELECT DATE_TRUNC('month', Order_Date) AS year_month, 
            ROUND(SUM(od.unit_price * od.quantity * (1-od.discount))::numeric,2) AS total_sales 
      FROM Orders o
      JOIN Order_Details od
        ON o.order_id = od.order_id
     GROUP BY DATE_TRUNC('month',Order_Date)
     ORDER BY year_month
    )
-- 'Using CTE monthly sales and getting the running total sales
SELECT *, ROUND(SUM(total_sales) OVER(order by year_month)::numeric,2) AS running_total
  FROM monthly_sales
```

From the running total Northwind does 1.2 million in sales in just under two years.

## Northwind Month-over-Month Growth Rate

Now that managment has the monthly sales they want to know month to month growth rate to help identify trends.

```python
%%sql
    WITH
    -- 'Grouping data into year and months then getting total sale for each month'
    monthly_sales AS
    (
    SELECT DATE_TRUNC('month', Order_Date) AS year_month, 
           ROUND(SUM(od.unit_price * od.quantity * (1-od.discount))::numeric,2) AS total_sales 
      FROM Orders o
      JOIN Order_Details od
        ON o.order_id = od.order_id
     GROUP BY DATE_TRUNC('month',Order_Date)
     ORDER BY year_month
    ),

    -- 'Creating and adding last month sales column to monthly_sales table'
    last_monthy_sales AS
    (
    SELECT *, LAG(total_sales) OVER(ORDER BY year_month) AS last_month_sales
      FROM monthly_sales
    )

    -- 'Getting the month over month growth rate'
    SELECT *, ROUND((total_sales - last_month_sales)/last_month_sales * 100,2) AS growth_rate
      FROM last_monthy_sales
```

Looking at the month to month growth, it shows that overall the growth rate is positive.  When the growth rate is negative it takes some time for the total sales to recuperate even though their is a positive growth rate.  February, June, August, and November of 1997 are the months with negative growth rates so it would be recommended that Northwind looks deeper into these months to see why.

## Identifying High-Value Customers

Now Northwind wants to get a better understanding of their customers and find out who are their high-value customers to offer special promotions.

```python
%%sql
    WITH
    -- 'Getting customer id, order id and total sales'
    customer_orders AS
    (
        SELECT o.customer_id, o.order_id, ROUND(SUM(unit_price * quantity * (1 - discount))::numeric,2) AS total_sales
          FROM orders o
          JOIN order_details od
            ON o.order_id = od.order_id
         GROUP BY o.customer_id, o.order_id
         ORDER BY o.order_id
    )

    -- 'Creating a case to determine if customer total sales are above average or below.'
    SELECT *,
    CASE
        WHEN total_sales > (SELECT AVG(total_sales) FROM customer_orders) THEN 'Above Average'
        ELSE 'Average/Below Average'
      END AS order_value
     FROM customer_orders
    LIMIT 5;

    
```

Now that every order has been labeled, it now makes it possible to find out which customers are your high-value customers.

```python
%%sql
    WITH
   -- 'Getting customer id, order id and total sales'
    customer_orders AS
    (
        SELECT o.customer_id, o.order_id, ROUND(SUM(unit_price * quantity * (1 - discount))::numeric,2) AS total_sales
          FROM orders o
          JOIN order_details od
            ON o.order_id = od.order_id
         GROUP BY o.customer_id, o.order_id
         ORDER BY o.order_id
    ),

    -- 'Determining if orders are above or below average using case'
    customer_order_value AS
    (
    SELECT *,
    CASE
        WHEN total_sales > (SELECT AVG(total_sales) FROM customer_orders) THEN 'Above Average'
        ELSE 'Average/Below Average'
    END AS order_value
    FROM customer_orders
    )

    -- 'Counting the number of above average orders each customer has'
    SELECT customer_id, COUNT(order_value) AS total_above_average_orders
      FROM customer_order_value
     WHERE order_value in ('Above Average')
     GROUP BY customer_id
     ORDER BY total_above_average_orders DESC
     LIMIT 10;
```

The top 3 customers by far have more above average orders then the rest of the customers.  Sending these customers special offers based on their order values can help with customer retention and help drive sales.  It can also help bring in new customers if they are sent a promotion for every new customer they refer.

## Percentage of Sales for Each Category

Northwind wants to get a better understanding of how the sales composition across different product categories looks like. They want to see which categories drive the most sales for the company and come up with marketing strategies.

```python
%%sql
WITH
    -- 'Getting category name and total sales'
    category_sales AS
    (
    SELECT c.category_name, ROUND(SUM(od.unit_price * od.quantity * (1-od.discount))::numeric,2) AS total_sales
      FROM order_details od
      JOIN products p
        ON od.product_id = p.product_id
      JOIN categories c
        ON p.category_id = c.category_id
     GROUP BY c.category_name
     ORDER BY total_sales desc
    )

    -- 'Getting percentage for each category based on total sales'
    SELECT *, 
           ROUND((total_sales / (SELECT SUM(total_sales) FROM category_sales) * 100)::numeric,2) AS percentage_total_sales
      FROM category_sales
```

The leading products for Northwind are Beverages, Dairy Products and Confections. They make up more then half of the sales for Northwind.  Grains/Cereals and Produce contribute the least and Northwind should take a closer look and make adjustments to these.

## Top Products Per Category

Lastly management wants to know their top performing products for each category to ensure which products to stock and promote.

```python
%%sql
    WITH
    -- 'Getting category, product name and total sales for each product'
    product_sales AS
    (
    SELECT c.category_name, p.product_name, 
           ROUND(SUM(od.unit_price * od.quantity * (1-od.discount))::numeric,2) AS total_sales
      FROM order_details od
      JOIN products p
        ON od.product_id = p.product_id
      JOIN categories c
        ON p.category_id = c.category_id
     GROUP BY p.product_name, c.category_name
     ORDER BY category_name, total_sales DESC
    )

    -- 'Ranking each product by category and looking at the top 3 ranking products for each category'
    SELECT *
      FROM (SELECT *, 
                   ROW_NUMBER() OVER(PARTITION BY category_name ORDER BY total_sales DESC) AS rank 
              FROM product_sales) AS ps
    WHERE rank <= 3
```

Looking at the top 3 products for each category, the number 1 rank products are contributing heavy for most categories.  Looking at the top beverage product it has almost 7 times as much sales then number 2 rank beverage. The other noticeable product is from Grains/Cereals and its the rank 3 product.  It has less then $10,000 in total sales making it a very low contributer.  Northwind should take the time to reevaluate the Grains/Cereals product line to increase its contribution to companies sales.
